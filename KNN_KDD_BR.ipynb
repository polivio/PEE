```python
# https://www.kaggle.com/c/kddbr-2019/data

# KNN KDD BR -Polivio
import numpy as np
from sklearn import preprocessing, neighbors
import pandas as pd
from sklearn.model_selection import train_test_split



```

```python
# preprocessing # blob # raw
train = pd.read_csv('https://github.com/polivio/PEE/raw/master/Training_grouped.csv')
test = pd.read_csv('https://github.com/polivio/PEE/raw/master/Test_grouped.csv')

#df.replace('?', -99999, inplace= True)
#df.drop(['id'] ,1, inplace= True)

train.head()
```

```python

train = train.loc[:,['traininglabelorder.scatterplotID', 'qtde.ponto', 
       'mean.signalX', 'sd.signalX', 'kurtosis.signalX',
      'mean.signalY', 'sd.signalY', 
       'kurtosis.signalY', 'L', 'X', 'XY', 'Y',
       'mean.silhouette', 'traininglabelorder.score']]



train.head(20)
```

```python
test = test.loc[:,['testorder.ID', 'qtde.ponto', 
       'mean.signalX', 'sd.signalX', 'kurtosis.signalX',
      'mean.signalY', 'sd.signalY', 
       'kurtosis.signalY', 'L', 'X', 'XY', 'Y',
       'mean.silhouette']]
test.head(20)
```

```python

# https://stackoverflow.com/questions/49420427/pandas-df-fillnadf-mean-not-working-on-multiindex-dataframe

train = train.reset_index()

train = train.fillna(0)
train.head(20)


```

```python
train.columns

```

```python
test = test.reset_index()

test = test.fillna(0)
test.head(20)
```

```python
# change the dtype to 'float64' 
train = train.astype('float64')
```

```python

X = np.array(train.drop(['traininglabelorder.score'], 1))
y = np.array(train['traininglabelorder.score'])


print(y)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Qual  o parmetro correto para random_state? e o significado de shuffle?

#X_test = np.array(test.drop(['traininglabelorder.score'], 1))
#y_test = np.array(test['? ]) --> primeira tentativa somente com os dados de treinamento.
```

```python
# classsifiers KNN

clf = neighbors.KNeighborsRegressor(n_neighbors=4)
clf.fit(X_train , y_train)
```

```python


accuracy = clf.score(X_test, y_test)

print(accuracy)
```
