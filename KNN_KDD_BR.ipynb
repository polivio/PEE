```python
# change the dtype to 'float64' 
train = train.astype('float64')
```

```python

X = np.array(train.drop(['traininglabelorder.score'], 1))
y = np.array(train['traininglabelorder.score'])


print(y)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Qual é o parâmetro correto para random_state? e o significado de shuffle?

#X_test = np.array(test.drop(['traininglabelorder.score'], 1))
#y_test = np.array(test['? ]) --> primeira tentativa somente com os dados de treinamento.
```

```python
# classsifiers KNN

clf = neighbors.KNeighborsRegressor(n_neighbors=4)
clf.fit(X_train , y_train)
```

```python


accuracy = clf.score(X_test, y_test)

print(accuracy)
```
